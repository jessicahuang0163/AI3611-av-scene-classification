```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
video_f1 = [0.51, 0.76, 0.78, 0.82, 0.87, 0.55, 0.63, 0.57, 0.72, 0.54]
audio_f1 = [0.74, 0.74, 0.66, 0.66, 0.86, 0.63, 0.63, 0.64, 0.82, 0.67]
mixed_f1 = [0.76, 0.88, 0.85, 0.87, 0.92, 0.69, 0.78, 0.73, 0.89, 0.74]
```

```python
index = np.arange(10)
bar_width = 0.2
fig, ax = plt.subplots(1, 1, figsize=(16, 5))
ax.bar(index - bar_width, video_f1, bar_width, label='Video')
ax.bar(index, audio_f1, bar_width, label='Audio')
ax.bar(index + bar_width, mixed_f1, bar_width, label='Mixed')

ax.set_xticks(index)
ax.set_xticklabels(['airport', 'bus', 'metro', 'metro_station', 'park', 'public_square', 'shopping_mall', 'street_pedestrian', 'street_traffic', 'tram'])
ax.set_ylabel('F1-score')
ax.set_xlabel('Classes')
ax.set_title('F1-score of different classes')
ax.legend()
fig.savefig("f1-score.pdf", dpi=400, bbox_inches='tight')
```
